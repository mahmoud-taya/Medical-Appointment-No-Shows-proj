```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
    # get month and day name and hour from Start Time after convert
df['Appointment_month'] = df['AppointmentDay'].dt.month
df['ppointment_day'] = df['AppointmentDay'].dt.day_name()
```

```python
#   cleaned, perform those cleaning steps in the second part of this section.
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
    # get month and day name and hour from Start Time after convert
df['register_month'] = df['ScheduledDay'].dt.month
df['rgister_day_name'] = df['ScheduledDay'].dt.day_name()
df['register_hour'] = df['ScheduledDay'].dt.hour
```

```python
# Wrong AppointmentDay
for d in df.ScheduledDay:
    for a in df.AppointmentDay:
        if d<=a:
            df['valid_Appointment']='wrong'
        else:
            df['valid_Appointment']=a-d 
            
```
